```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Sample data
students = [f"Student_{i+1}" for i in range(30)]
subjects = ['Maths', 'Physics', 'Chemistry', 'CS', 'Biology']
subject_type = {
    'Maths': 'Theory',
    'Physics': 'Theory',
    'Chemistry': 'Lab',
    'CS': 'Lab',
    'Biology': 'Theory'
}
dates = [datetime(2025, 8, 1) + timedelta(days=i) for i in range(5)]

# Generate dataset
data = []
for date in dates:
    for student in students:
        for subject in subjects:
            present = random.choice(['Present', 'Absent'])
            data.append({
                'Date': date.strftime('%Y-%m-%d'),
                'Student Name': student,
                'Subject': subject,
                'Subject Type': subject_type[subject],
                'Attendance': present
            })

df = pd.DataFrame(data)

# Calculate Total Attendance for each student
attendance_summary = (
    df[df['Attendance'] == 'Present']
    .groupby('Student Name')
    .size()
    .reset_index(name='Total Present')
)

# Merge into main DataFrame
df = df.merge(attendance_summary, on='Student Name', how='left')
df['Total Present'].fillna(0, inplace=True)
df['Total Present'] = df['Total Present'].astype(int)

# Save to CSV
csv_path = "/content/student_attendance_dataset.csv"
df.to_csv(csv_path, index=False)

print(f"✅ Dataset created with {len(df)} rows and saved to:\n{csv_path}")
```
